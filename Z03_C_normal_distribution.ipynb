# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
import scipy.stats
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
from scipy.stats import multivariate_normal

mu_x = 0
mu_y = 0
#Create grid and multivariate normal
x = np.linspace(-3,3,1000)
y = np.linspace(-3,3,1000)
X, Y = np.meshgrid(x,y)
# dziedzina sposob pierwszy
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y

rv1 = multivariate_normal([mu_x, mu_y], [[1, 0], [0, 1]])
rv2 = multivariate_normal([mu_x, mu_y], [[2, 0], [0, 1]])
rv3 = multivariate_normal([mu_x, mu_y], [[1, 0], [0, 2]])

#Make a 3D plot
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1, 3, 1, projection='3d', title='[[1,0],[0,1]]')
ax2 = fig.add_subplot(1, 3, 2, projection='3d', title='[[2,0],[0,1]]')
ax3 = fig.add_subplot(1, 3, 3, projection='3d', title='[[1,0],[0,2]]')
ax1.plot_surface(X, Y, rv1.pdf(pos),cmap='viridis',linewidth=0)
ax2.plot_surface(X, Y, rv2.pdf(pos),cmap='viridis',linewidth=0)
ax3.plot_surface(X, Y, rv3.pdf(pos),cmap='viridis',linewidth=0)

plt.show()
```

# Zad 
Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
import numpy as np
from scipy.stats import multivariate_normal

#Parameters to set
mu_x = 0
mu_y = 0

rv1 = multivariate_normal([mu_x, mu_y], [[1, 0], [0, 1]])
rv2 = multivariate_normal([mu_x, mu_y], [[2, 0], [0, 1]])
rv3 = multivariate_normal([mu_x, mu_y], [[1, 0], [0, 2]])
# dziedzina sposob drugi
x = np.linspace(-3,3,500)
y = np.linspace(-3,3,500)
X,Y = np.meshgrid(x,y)
pos = np.array([X.flatten(),Y.flatten()]).T

fig = plt.figure(figsize=(30,10))
ax4 = fig.add_subplot(131, title='[[1,0],[0,1]]')
ax5 = fig.add_subplot(132, title='[[2,0],[0,1]]')
ax6 = fig.add_subplot(133, title='[[1,0],[0,2]]')
ax4.contour(X, Y, rv1.pdf(pos).reshape(500,500))
ax5.contour(X, Y, rv2.pdf(pos).reshape(500,500))
ax6.contour(X, Y, rv3.pdf(pos).reshape(500,500))
ax4.sharey(ax5)

plt.show()
```

# Zad 
Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

i narysuj odpowiedni wykres

```python
#Parameters to set
mu_x = 0
mu_y = 0
N=10000
rv1 = multivariate_normal([mu_x, mu_y], [[1, 0], [0, 1]])
rv2 = multivariate_normal([mu_x, mu_y], [[2, 0], [0, 1]])
rv3 = multivariate_normal([mu_x, mu_y], [[1, 0], [0, 2]])

sample1 = rv1.rvs(size=N)
sample2 = rv2.rvs(size=N)
sample3 = rv3.rvs(size=N)


fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15,5), sharey=True)
ax1.scatter(sample1[:,0], sample1[:,1])
ax2.scatter(sample2[:,0], sample2[:,1])
ax3.scatter(sample3[:,0], sample3[:,1])

plt.show()

```

```python
a=0.5
rv4 = multivariate_normal([mu_x, mu_y], [[1,a], [a,5]])
sample4 = rv4.rvs(size=N)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.scatter(sample4[:,0], sample4[:,1])
plt.show()
```
