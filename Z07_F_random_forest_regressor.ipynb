# Random Forest Regressor

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

Proszę narysować wykres dowolnej funkcji na odcinku $[0,3]$.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
x = stats.uniform(0,3).rvs(100)
y = f(x) + stats.norm(0,0.2).rvs(len(x))
```

# Zadanie
wykonaj regresję z różnymi parametrami

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 1000))
regr_rf = RandomForestRegressor(max_depth=5, n_estimators =5 ,random_state=2)
y_rbf = regr_rf.fit(x, y)
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, regr_rf.predict(x_plot), color='blue',linewidth=3)
plt.show()
```

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 1000))
regr_rf = RandomForestRegressor(max_depth=10, max_features=1, n_estimators=10, random_state=3)
y_rbf = regr_rf.fit(x, y)
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, regr_rf.predict(x_plot), color='blue',linewidth=3)
plt.show()
```

```python
for n in range(1,6):
    regr_rf = RandomForestRegressor(max_depth=5, n_estimators = n ,random_state=2)
    regr_rf.fit(x,y)
    axes = plt.gca()
    axes.set_xlim([0,3])
    axes.set_ylim([0,8])
    plt.plot(x_plot, regr_rf.predict(x_plot), linewidth=3)
plt.show()
```

# Zad
Wyznacz optymalne parametry w modelu RandomForestRegressor za pomocą GridSearchCV

```python
rfr = GridSearchCV(RandomForestRegressor(random_state=3), cv=5,
        param_grid={"max_depth": range(1,16),
        "n_estimators": range(1,16)},
        refit = True)
rfr.fit(x,y)
rfr.best_params_
```

# Zad

Wyznacz optymalne parametry dla modelu RandomForestRegressor.

```python
import pandas as pd
df_adv = pd.read_csv('https://raw.githubusercontent.com/przem85/podstawy_sztucznej_inteligencji/main/Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
from sklearn import  metrics
rfr = GridSearchCV(RandomForestRegressor(random_state=3), cv=5,
        param_grid={"max_depth": range(1,16),
        "n_estimators": range(1,16)},
        refit = True)
rfr.fit(X,y)
rfr.best_params_
print("R^2: {}".format(metrics.r2_score(y, rfr.predict(X))))
```
